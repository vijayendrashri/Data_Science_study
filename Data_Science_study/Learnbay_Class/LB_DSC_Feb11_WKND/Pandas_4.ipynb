### Groupby 

```python
import pandas as pd
```

```python
df = pd.read_csv('dataset/weather_by_cities.csv')
df
```

```python
df['city'].unique()
```

```python
g = df.groupby('city')
```

```python
for city,data in g:
    print(city)
    print(data)
```

## Merge

```python
df1 = pd.DataFrame({
"city": ["new york","chicago","orlando"],
"temperature": [21,14,35],
})
df1 
```

```python
df2 = pd.DataFrame({
"city": ["chicago","new york","orlando"],
"humidity": [65,68,75],
})
df2 
```

```python
df3=pd.merge(df1,df2,on='city')
```

```python
df3
```

```python
df1 = pd.DataFrame({
"city": ["new york","chicago","orlando", "baltimore"],
"temperature": [21,14,35, 38],
})
df1 
```

```python
df2 = pd.DataFrame({
"city": ["chicago","new york","san diego"],
"humidity": [65,68,71],
})
df2 
```

```python
pd.merge(df1,df2,on='city',how='left')
```

```python
pd.merge(df1,df2,on='city',how='right')
```

```python
pd.merge(df1,df2,on='city',how='outer')
```

```python
pd.merge(df1,df2,on='city',how='outer',indicator=True,sort=True,)
```

```python
df1 = pd.DataFrame({
"city": ["new york","chicago","orlando", "baltimore"],
"temperature": [21,14,35, 38],
 "humidity": [65,68,71,45]
})

df2 = pd.DataFrame({
"city": ["chicago","new york","san diego","baltimore"],
"temperature": [21,14,35, 38],
"humidity": [65,68,71,56]
})
```

```python
pd.merge(df1,df2,on='city',suffixes=('_first','_second'))
```

## Pivot
### Tranform and reshape data

```python
df = pd.read_csv('dataset/weather.csv')
df
```

```python
# Transform index to column
df.pivot(index='date',columns='city')
```

```python
df.pivot(index='city',columns='date',values='humidity')
```

## Pivot Table
#### Its use for summerize and agg data

```python
df = pd.read_csv('dataset/weather2.csv')
df
```

```python
df.pivot_table(index='city',columns='date')
```

```python
df.pivot_table(index='city',columns='date',aggfunc='max')
```

## Magin

```python
df.pivot_table(index='city',columns='date',aggfunc='max',margins=True)
```

```python
df.pivot_table(index='city',columns='date',aggfunc='mean',margins=True)
```

## Stacking and unstacking

```python
df =pd.read_csv('dataset/avocado.csv')
df
```

```python
df12 =df.pivot(index='Region',columns='Type')
```

```python
df12
```

```python
df_s=df12.stack()
df_s
```

```python
df_s.stack().loc['Atlanta']
```

```python
df_s.stack(level=0)
```

```python
# stack - when you convert a column to row
# unstack - when you convert a row to column
df_s.stack(level=0).unstack()
```

```python
df_s.unstack(level=1)
```
