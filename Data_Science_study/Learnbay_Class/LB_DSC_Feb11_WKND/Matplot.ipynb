```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
week_day=[1,2,3,4,5,6,7]
temp=[50,51,52,48,47,46,49]
```

```python
plt.plot(week_day,temp)
```

```python
plt.plot(week_day,temp,color='green')
```

```python
plt.plot(week_day,temp,color='green',marker='o',linestyle='--')
```

```python
plt.plot(week_day,temp,color='green',marker='o',linestyle='--')
plt.grid()
```

```python
plt.plot(week_day,temp,'bv--')
plt.grid()
```

```python
plt.plot(week_day,temp,'vb--')
plt.grid()
```

```python
plt.figure(figsize=(20,20))
plt.plot(week_day,temp,'bv--')
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
plt.grid()
```

```python
#Marker Size
plt.plot(week_day,temp,'bv--',markersize=10)
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
```

```python
#Alpha
plt.plot(week_day,temp,'vb--',alpha=0.4)
plt.xlabel('Day')
plt.ylabel('Temp')
plt.title('Weather')
```

```python
days=[1,2,3,4,5,6,7] # weekdays
max_t=[50,51,52,48,47,49,46] # max temp recorder at that day
min_t=[43,42,40,44,33,35,37] # min temp
avg_t=[45,48,48,46,40,42,41] # avg temp 
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')
plt.legend()
plt.xlabel('day')
plt.ylabel('temp')
plt.title('Weather')
plt.grid()
```

```python
plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label='min')
plt.plot(days,avg_t,label='avg')

plt.xlabel('day')
plt.ylabel('temp')
plt.title('Weather')
plt.grid()
plt.legend(loc='right',fontsize='large',shadow = True)
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27] 
profit=[100,99,90,88]
```

```python
xpos=np.arange(len(company))
plt.bar(xpos+.4,revenue,label='Rev',width=0.3)
plt.bar(xpos-.4,profit,label='Prof',width=0.3)
plt.xticks(xpos)
plt.legend()
```

```python
plt.barh(company,revenue,label='Rev')
plt.legend()
```

```python
year=[2014,2015,2016,2017]
income=[4000,4500,5300,4600]
expense=[2800,3000,2800,3400] 

plt.subplot(2,1,1)
plt.bar(year,income,color='g')
plt.subplot(2,2,4)
plt.bar(year,expense,color='b')
plt.subplot(2,2,3)
plt.scatter(year,expense,color='b')
plt.savefig('multi.pdf')
```
