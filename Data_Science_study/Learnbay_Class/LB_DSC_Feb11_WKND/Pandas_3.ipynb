# Iteration in Pandas

1. tuples
2. rows
3. items

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
'Headquarters' : ["Paris", "Bangalore", "Washington",
"Tokyo", "Beijing"],
'Country_name' : ["France", "India", "USA", "Japan", "China"],
'Formation_year' : [1961, 1969, 1958, 2003, 1993]
} 
```

```python
df = pd.DataFrame(space_org)
df
```

```python
for rows in df.itertuples():
    print(rows)
```

```python
for rows in df.itertuples():
    print(rows.Organizations_name)
```

```python
for rows in df.iteritems():
    print(rows)
```

```python
for key,value in df.iteritems():
    if key == 'Organizations_name':
        print(value)
```

```python
for rows in df.iteritems():
     print(rows[1][4])
```

```python
for rows in df.iterrows():
    print(rows)
```

```python
for rows in df.iterrows():
    print(rows[1][0])
```

```python
for rows in df.iterrows():
    print(rows[1].Organizations_name)
```
