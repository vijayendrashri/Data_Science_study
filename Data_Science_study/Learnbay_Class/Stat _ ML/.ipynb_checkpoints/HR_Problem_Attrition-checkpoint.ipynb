```python
# Load all required libs.
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
# Load dataset in Pandas Object 
hr_df = pd.read_csv("HR_Employee_Attrition_Data.csv")
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
hr_df
```

```python
hr_df.describe(exclude=[np.object])
```

### Exploratory Data Analysis

```python
# Convert Attrition Yes/NO value to 1/0.
le = preprocessing.LabelEncoder()
hr_df.Attrition = le.fit_transform(hr_df.Attrition)
```

```python
# Over18 has only one unique value which represents all employees are above 18 years of age.
# Few have std=0
hr_df.drop(['EmployeeCount','StandardHours','Over18','EmployeeNumber'],axis=1,inplace=True)
```

```python
 # Get categorical feature names
cat_col_names = hr_df.select_dtypes(include=[np.object]).columns.tolist()
cat_col_names
```

```python
# Check Null value count
hr_df.isnull().sum()
```

### Exploratory Data Analysis - Categorical Features

```python
def generate_frequency_df(df,cols_name):
    new_df=pd.DataFrame()
    percent_value = df.groupby(cols_name)['Attrition'].value_counts(normalize=True)
    new_df = percent_value.mul(100).rename('Percent').reset_index()
    return new_df
```

```python
# Get Percentage of Employee attrition with respect to each categorical features.
for cols in cat_col_names:
    new_df=generate_frequency_df(hr_df,cols)
    g = sns.catplot(x=cols, y='Percent', hue='Attrition', kind='bar', data=new_df)
    for p in g.ax.patches:
        txt = str(p.get_height().round(1)) + '%'
        g.ax.text(p.get_x(),p.get_height(),txt)
        labels = g.ax.get_xticklabels()
        g.ax.set_xticklabels(labels, rotation=45)
    g.ax.set_ylim(0,100)
# As below bar shows that high rate of Attrition for employee who Travel_Frequently,
# Sales,HR & Technical Degree,Male, Sales Representative, Single and OverTime.
```

### Exploratory Data Analysis - Ordered Categorical Features

```python
# Capture only ordinal features
ordinal_cols = ["Education","EnvironmentSatisfaction","JobInvolvement","JobLevel","JobSatisfaction","PerformanceRating","RelationshipSatisfaction","WorkLifeBalance","StockOptionLevel"]
```

```python
# Get Percentage of Employee attrition with respect to each Ordinal features.
for cols in ordinal_cols:
    new_df=generate_frequency_df(hr_df,cols)
    g = sns.catplot(x=cols, y='Percent', hue='Attrition', kind='bar', data=new_df)
    for p in g.ax.patches:
        txt = str(p.get_height().round(1)) + '%'
        g.ax.text(p.get_x(),p.get_height(),txt)
    g.ax.set_ylim(0,100)
# As below bar shows that high rate of Attrition for employee.
```

```python
# Filter only numeric columns and make Attrition as target column.
all_cols=hr_df.columns.to_list()
target=['Attrition']
numeric_cols=set(all_cols)-set(ordinal_cols)-set(cat_col_names)-set(target)
```

```python
numeric_cols_df = hr_df[numeric_cols]
numeric_cols_df.head()
```

### Let's create dummy variables for each categorical attribute for training our classification model

```python
def preprocessor(df,cat_col_names):
    df = pd.get_dummies(data = df, columns=cat_col_names)
    return df
```

```python
# Convert categorical to OneHotLabel
cat_col_df = hr_df[cat_col_names]
cat_df = preprocessor(cat_col_df,cat_col_names)
cat_df.shape
```

```python
# Convert Ordinal to OneHotLabel
ordinal_col_df = hr_df[ordinal_cols]
ordinal_df = preprocessor(ordinal_col_df,ordinal_cols)
ordinal_df.shape
```

```python
# Pull all target values
target_df=hr_df[target]
```

### Correlation of Numerical Features against Attrition

```python
corr_df = pd.concat([numeric_cols_df,target_df],axis=1)
corr = corr_df.corr().abs()
```

```python
plt.figure(figsize = (10, 8))
sns.axes_style("white")
sns.heatmap(data=corr, annot=True, square=True, linewidths=.5, vmin=-1, vmax=1, cmap="YlGnBu")
plt.show()
```

### Concat the two dataframes together columnwise

```python
final_df = pd.concat([numeric_cols_df,ordinal_df,cat_df],axis=1)
final_df.head()
```

```python
# Create standarization for every columns
scale =StandardScaler()
scale_data = scale.fit_transform(final_df)
print(scale_data)
```

```python
# Split dataset in Training and test set
x_train, x_test, y_train, y_test = train_test_split(final_df,target_df,test_size =0.2,random_state=24)
print("Stratified Sampling: ", len(x_train), "train set +", len(x_test), "validation set")
```

```python
x_train.DailyRate.mean()
```

```python
x_test.DailyRate.mean()
```

```python
# Model performance function
def gen_model_performance(actual_target, pred_target):
    model_roc_score = metrics.roc_auc_score(actual_target, pred_target)
    model_accuracy = metrics.accuracy_score(actual_target, pred_target)*100
    precision = metrics.precision_score(actual_target, pred_target)
    recall = metrics.recall_score(actual_target, pred_target)
    f1_score = metrics.f1_score(actual_target, pred_target)
    return model_roc_score,model_accuracy,precision,recall,f1_score
```

### Using GridSearchCV

```python
fit_dt = DecisionTreeClassifier()
```

np.random.seed(42)

                     param_grid=param_dist, 
                     n_jobs = 4)

cv_dt.fit(x_train, y_train)

print(dtc)

### Build the Decision Tree Model

```python
dtc = DecisionTreeClassifier(criterion='entropy', max_depth=16, min_samples_split=3,splitter='random')
model_dt = dtc
model_dt.fit(x_train, y_train)
model_dt_score_train = model_dt.score(x_train, y_train)
print("Training score: ",model_dt_score_train)
model_dt_score_test = model_dt.score(x_test, y_test)
print("Testing score: ",model_dt_score_test)
pred = model_dt.predict(x_test)
model_roc_score,accuracy,precision,recall,f1_score = gen_model_performance(y_test, pred)
```

## Performance Metrics

```python
model_results = pd.DataFrame(columns = ["Model", "Accuracy", "Precision","F1 Score","ROC Score"])
model_results = model_results.append({"Model": model_dt.__class__.__name__,
                              "Accuracy": accuracy, "Precision": precision,"F1 Score": f1_score,
                              "ROC Score": model_roc_score}, ignore_index=True)
model_results
```

```python
fpr_dt, tpr_dt, _ = metrics.roc_curve(y_test, pred)
roc_auc_dt = metrics.auc(fpr_dt, tpr_dt)
```

### Area Under the Curve

```python
plt.figure(1)
lw = 2
plt.plot(fpr_dt, tpr_dt, color='green',
         lw=lw, label='Decision Tree(AUC = %0.2f)' % roc_auc_dt)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Area Under Curve')
plt.legend(loc="lower right")
plt.show()
```
